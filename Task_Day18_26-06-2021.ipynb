```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/Social_Network_Ads.csv")
df
```

### Line Plot

```python
plt.title("Line Plot")
plt.plot(df['EstimatedSalary'])
plt.show()
```

### Scatter Plot

```python
plt.title("Scatter Plot")
plt.scatter(df["Age"],df["EstimatedSalary"])
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.show()
```

### Histogram

```python
plt.title("Histogram")
plt.hist(df["EstimatedSalary"])
plt.xlabel("EstimatedSalary")
plt.show()
```

### Bar Graph

```python
plt.title("Bar Graph")
plt.bar(df["Purchased"],df["Age"])
plt.xlabel("Purchased")
plt.ylabel("Age")
plt.show()
```

### Pie Chart

```python
plt.title("Pie Chart")
plt.pie(df["Age"][:10],labels = df["Age"][:10], autopct = "% 1.1f%%")
plt.show()
```

### Box Plot

```python
plt.title("Box Plot")
plt.boxplot(df["User ID"])
plt.show()
```

### Sub-Plots

```python
fig, area = plt.subplots(2, 2)
area[0][0].plot(df['Gender'])
area[0][1].plot(df['Purchased'])
area[1][0].plot(df['EstimatedSalary'])
area[1][1].plot(df['Age'])
plt.show()
```
