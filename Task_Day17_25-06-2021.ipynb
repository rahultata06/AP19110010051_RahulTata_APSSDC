```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv')
titanic
```

```python
titanic.isnull().sum()
```

### Replace null values 'Age' column with ffill()

```python
titanic['Age'].fillna(method = "ffill", inplace = True)
titanic
```

### Replace null values in 'Cabin and 'Embarked' columns with most occured value

```python
titanic['Cabin'].fillna(titanic['Cabin'].mode().iloc[0], inplace = True)
titanic['Embarked'].fillna(titanic['Embarked'].mode().iloc[0], inplace = True)
titanic
```

```python
titanic.isnull().sum()
```

### Convert Cabin, Embarked columns into numerical columns

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
titanic['Cabin'] = le.fit_transform(titanic['Cabin'])
titanic['Embarked'] = le.fit_transform(titanic['Embarked'])
titanic['Name'] = le.fit_transform(titanic['Name'])
titanic['Sex'] = le.fit_transform(titanic['Sex'])
titanic['Ticket'] = le.fit_transform(titanic['Ticket'])
titanic
```

### Display the plot of original data and robust scaler data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
advertisement=pd.read_csv("https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv")
advertisement.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
advertisement.isnull().sum()
```

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
re=pd.DataFrame(rs.fit_transform(advertisement),columns=advertisement.columns)
```

```python
fig,(axes1,axes2)=plt.subplots(1,2,figsize=(9,4))

sns.kdeplot(re["TV"],ax=axes1)
sns.kdeplot(re["radio"],ax=axes1)
sns.kdeplot(re["newspaper"],ax=axes1)
sns.kdeplot(re["sales"],ax=axes1).set(title="After Robust Scaling")

sns.kdeplot(advertisement["TV"],ax=axes2).set(title="Before Robust scaling")
sns.kdeplot(advertisement["radio"],ax=axes2)
sns.kdeplot(advertisement["newspaper"],ax=axes2)
sns.kdeplot(advertisement["sales"],ax=axes2)

plt.tight_layout()
```
